## 2. Модифицируйте код из предыдущей работы так, чтобы модель регрессии умела работать с входными данными, состоящими из произвольного количества столбцов.

```python
class Model(object):
    """Модель парной линейной регрессии"""
    
    def __init__(self):
        self.b0 = 0
        self.b1 = 0
        
    def predict(self, X, x_col=0):
        if str(type(X))=="<class 'pandas.core.frame.DataFrame'>":
            X=X.iloc[:,x_col]
            
        return self.b0 + self.b1 * X
    
    def error(self, X, Y, x_col=0,y_col=0):
        if str(type(X))=="<class 'pandas.core.frame.DataFrame'>":
            X=X.iloc[:,x_col]
        if str(type(Y))=="<class 'pandas.core.frame.DataFrame'>":
           Y=Y.iloc[:,y_col]
           
        s = sum(((self.predict(X) - Y)**2) / (2 * len(X)))

        return s
    
    def fit(self, X, Y, alpha=1, accuracy=0.01, max_steps=10000,til=1e-6, x_col=0,y_col=0):
        if str(type(X))=="<class 'pandas.core.frame.DataFrame'>":
            X=X.iloc[:,x_col]
        if str(type(Y))=="<class 'pandas.core.frame.DataFrame'>":
            Y=Y.iloc[:,y_col]
                
        steps, errors = [], []
        step = 0
        
        for _ in range(max_steps):
            dJ0 = sum(self.predict(X) - Y) /len(X)
            dJ1 = sum((self.predict(X) - Y) * X) /len(X)
            self.b0 -= alpha * dJ0
            self.b1 -= alpha * dJ1    
            new_err = self.error(X, Y)
            step += 1            
            steps.append(step)
            errors.append(new_err)
            
            if len(errors)>=2:
                if (errors[-2]-errors[-1])<til:
                    print('Ошибка перестала существенно меняться')
                    break
                
                elif (errors[-2]-errors[-1])>0:
                    alpha=alpha/2
                   
            elif len(errors)==max_steps:
                print('Все шаги пройдены')
        
        return steps, errors
    
    def plot(self,X, Y, x_col=0,y_col=0):
        if str(type(X))=="<class 'pandas.core.frame.DataFrame'>":
            X=X.iloc[:,x_col]
        if str(type(Y))=="<class 'pandas.core.frame.DataFrame'>":
            Y=Y.iloc[:,y_col]
            
        xmin=(min(X))
        xmax=(max(X))
        X0 = np.linspace(xmin, xmax, 100)
        Y0 = self.predict(X0)
        plt.figure()
        plt.scatter(X, Y)
        plt.plot(X0, Y0, 'r')
        plt.show()
    
    def score(self,X,Y, x_col=0,y_col=0, prin=True):
        if str(type(X))=="<class 'pandas.core.frame.DataFrame'>":
            X=X.iloc[:,x_col]
        if str(type(Y))=="<class 'pandas.core.frame.DataFrame'>":
            Y=Y.iloc[:,y_col]
        results={}
        
        y_pred = self.predict(X)
        y_mean = sum(Y)/len(Y)
        
        ssreg = sum([(y_predi-y_mean)**2 for y_predi in y_pred])
        sstot = sum([(yi - y_mean)**2 for yi in Y])
        results['Determination'] = ssreg / sstot
        
        smape =  sum([abs(yu[1] - yu[0])/yu[1] for yu in list(zip(y_pred,Y))])/len(Y)
        results['1 - MAPE'] = 1 - smape
        
        if prin:
            for key,value in results.items():
                print(f'{key} = {value}')
        
        
        
    


for i in range(x.shape[1]):
    #display(x[[i]])
    binary = Model()
    steps, errors = binary.fit(x[[i]], y)
    binary.predict(x[[i]])
    binary.plot(x[[i]], y)
    binary.score(x[[i]], y)
    print(f'Error after gradient descent: = {binary.error(x[[i]],y)}')
```

## 3. Для целевой переменной и каждого из признаков постройте полиномиальную регрессию второго, третьего и десятого порядков. Сделайте вывод о значимости регрессии.

```python
dictx=dict()
degrees=[1,2,3,10,20,25,26, 27]
for i in range(x.shape[1]):
    print(f'\n\n\nX{i}')
    dictx[f'X{i}']=[]
    for deg in degrees:
        print(f'\nDegree = {deg}')
        xp=np.array(x[[i]])
        if deg!=1:
            pf = PolynomialFeatures(degree=deg)
            xp = pf.fit_transform(xp)
        
        binary = LinearRegression()
        binary.fit(xp, y)
        
        binary.predict(xp)

        if deg==1:
            plt.scatter(xp, y)
            xx = np.linspace(xp.min(), xp.max(), 100).reshape((-1, 1))
            plt.plot(xx, binary.predict(xx), c='r')
        plt.show()
        print(f'Score = {binary.score(xp,y)}')
        dictx[f'X{i}'].append(binary.score(xp,y))
        print(f'Intercept = {binary.intercept_}')
print(dictx)
```
