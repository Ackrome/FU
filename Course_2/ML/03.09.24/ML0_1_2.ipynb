**Join Merge**

```python
df1 = pd.DataFrame({'A': ['5', '87', '42', '43'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']}, index=['i1', 'i2', 'i3', 'i4'])

df2 = pd.DataFrame({'A1': ['43', '3', '80', '5'],
                     'B1': ['B4', 'B5', 'B6', 'B7'],
                     'C1': ['C4', 'C5', 'C6', 'C7'],
                     'D1': ['D4', 'D5', 'D6', 'D7']}, index=['i1', 'i4', 'i5', 'i2'])
```

```python
df1.join(df2, how='outer') #how = left, right, inner, outer
```

```python
df1.merge(df2, left_on='A', right_on='A1')
```

**concat**

```python
pd.concat([df1, df2], axis=1)
```
