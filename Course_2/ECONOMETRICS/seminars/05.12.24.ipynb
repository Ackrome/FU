## Модель с фективными пе

```python
import numpy as np
import pandas as pd
from statsmodels.api import OLS, add_constant

x2 = 'CABABBACCACAABCBCACBBACCBACBCABBCAAB'
x1 = '29 53 29 58 66 67 63 59 51 67 63 33 33 42 67 33 23 28 19 30 23 21 56 45 43 38 37 43 27 43 45 48 47 48 53 58'
y = '36 69 47 73 64 60 62 71 62 70 71 52 63 48 71 46 25 55 28 40 41 56 62 50 45 58 46 58 34 65 55 57 59 64 61 62'

# Преобразование данных
x2 = np.array([i for i in x2])
x1 = np.array(x1.split(' ')).astype(int)
y = np.array(y.split(' ')).astype(int)

# Создание фиктивных переменных
x_dummies = pd.get_dummies(x2, prefix="x", drop_first=True)

# Подготовка данных для регрессии
X = pd.DataFrame({'x1': x1})
X = pd.concat([X, x_dummies], axis=1)

# Преобразование в числовой формат
X = X.astype(float)


# Модель регрессии
model = OLS(y, add_constant(X)).fit()

# Результаты
model.summary()
```
