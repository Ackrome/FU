```python
X=np.arange(10,100)
X=X[list(map(isprime,X))]
```

```python
print('Среднее геометрическое G(X): ', st.gmean(X))
print('Среднее гармоническое H(X): ', st.hmean(X))
# Вычисление по формуле
print('Среднее гармоническое H(X): ', len(X)/sum(1/X))
print('Среднее степенное Q(X) c показателем 3: ', (sum(X**3)/len(X))**(1/3))
```

```python
X=np.arange(1,101)
S=np.sin(X)
C=np.cos(X)
```

```python
S.mean(),C.mean()
```

```python
S.std(),C.std()
```

```python
st.moment(S,4), st.moment(C,4)
```

```python
st.kurtosis(S),st.kurtosis(C)
```

```python
np.corrcoef(S,C)
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
 
ecdf = ECDF(S)
y = ecdf(S)
S.sort()
y.sort()

ecdf = ECDF(C)
yy = ecdf(C)
C.sort()
yy.sort()

fig,ax =plt.subplots(figsize=(25, 10))
plt.tick_params(labelsize = 20)
plt.tick_params(labelsize = 20)
plt.title("График эмпирической функции распределения совокупности sin.")
plt.xlabel('$x$')
plt.step(S,y,color = 'r', label = 'sin')
plt.step(C,yy,color = 'b', label = 'sin')
```
