Пусть $\overrightarrow{X}=(X_1,...,X_n)$ - выборка объёма $n$ из распределения, зависящего от некоторого числового параметра $\theta\in\Theta$.


 Случайный промежуток $I(\overrightarrow{X})$ называется __$\gamma$-доверительным интервалом__ для параметра $\theta$, если 
 
 $$P\{\theta\in I(\overrightarrow{X})\}=\gamma, \qquad\forall \theta\in\Theta.$$

__Решение.__

<font color="red"> __а) Доверительный интервал для генерального среднего при известной дисперсии__ $\sigma^2$:</font>

$$(\overline{X}-\delta;\overline{X}+\delta), $$

где
$\delta=\frac{z_{\frac{1+\gamma}2}\cdot\sigma}{\sqrt{n}}$, $\quad z_{(1+\gamma)/2}$ - квантиль  $N(0;1)$  уровня $(1+\gamma)/2$.

```python
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
import matplotlib.ticker as ticker
```

```python
gamma = 0.96
sigma = 0.1
```

```python
N = 10**5

mu = 2

data = sts.norm.rvs(loc=mu, scale=sigma, size=N)
n = 10**3

sample_means=[]
intervals = []
b=[]
t=[]
m=[]

for sample in range(50):
    sample = np.random.choice(a=data, size=n)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    z_gamma = sts.norm.ppf((1+gamma)/2) 
    
    delta = z_gamma * sigma/ n**0.5

    confidence = (sample_mean - delta, sample_mean + delta) 
    intervals.append(confidence)
    
for bot,top in intervals:
    b.append(bot)
    t.append(top)
    m.append((top-bot)/2)
    

    
fig,ax =plt.subplots(figsize=(10, 10))
plt.tick_params(labelsize = 30)
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.006))
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
plt.tick_params(labelsize = 10)
plt.grid(color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=1)

for i in range(len(intervals)):
    if b[i] < np.mean(data)<t[i]:
        plt.errorbar(y=i, x=sample_means[i], xerr=m[i],fmt='-o',ecolor='g',capsize=10)
    else:
        plt.errorbar(y=i, x=sample_means[i], xerr=m[i],fmt='-o',ecolor='r',capsize=10)

plt.vlines(ymin=0, ymax=50, x=np.mean(data), linewidth=2.0,color="red")
plt.show()
```
