```python
from sympy import *
from scipy.stats import *
import pandas as pd
import numpy as np
init_printing(use_unicode=True,use_latex=True)
```

```python
x,y,C = symbols('x y C', real = True)
f = C*(x**2+y**2)
C0=solve(integrate(integrate(f,(y,0,1)),(x,0,1))-1,C)[0]
f=f.subs(C,C0)

f_x=integrate(f,(y,0,1))
f_y=integrate(f,(x,0,1))

EX=integrate(x*f_x,(x,0,1))
EY=integrate(y*f_y,(y,0,1))

EX2=integrate(x**2*f_x,(x,0,1))
EY2=integrate(y**2*f_y,(y,0,1))

VARX=EX2-EX**2
VARY=EY2-EY**2

sig_x=VARX**(0.5)
sig_y=VARY**(0.5)

EXY=integrate(integrate(x*y*f,(y,0,1)),(x,0,1))

COVXY=EXY-EX*EY

RHOXY=COVXY/(sig_x*sig_y)

C0, EX, EY, VARX,VARY,sig_x,sig_y,COVXY,RHOXY
```

```python
x,y,C = symbols('x y C', real = True, )
f = C*x*y

x_bound = [x,0,1]
y_bound_nx = [y,0,1]

C0=solve(integrate(f,[y,0,-x+1],[x,0,1])-1,C)[0]
f=f.subs(C,C0)

f_x=integrate(f,[y,0,-x+1])
f_y=integrate(f,[x,0,-y+1])

EX=integrate(x*f_x,(x,0,1))
EY=integrate(y*f_y,(y,0,1))

EX2=integrate(x**2*f_x,(x,0,1))
EY2=integrate(y**2*f_y,(y,0,1))

VARX=EX2-EX**2
VARY=EY2-EY**2

sig_x=sqrt(VARX)
sig_y=sqrt(VARY)

EXY=integrate(integrate(x*y*f,[y,0,-x+1]),(x,0,1))

COVXY=EXY-EX*EY

RHOXY=COVXY/(sig_x*sig_y)

ruz = pd.DataFrame(np.array('C0,EX,EY,VARX,VARY,sig_x,sig_y,COVXY,RHOXY'.split(',')))

ruz[1] = np.array([C0, EX, EY, VARX,VARY,sig_x,sig_y,COVXY,RHOXY])
ruz
```
