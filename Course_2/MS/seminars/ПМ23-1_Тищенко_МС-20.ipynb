```python
import numpy as np
from scipy.stats import *
from sympy import *
from IPython.display import Math,Latex
from scipy import optimize
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

def rrstr(x,n): # округление до n знаков после запятой
    fmt = '{:.'+str(n)+'f}'
    return fmt.format(x).replace('.',',')

v_rrstr = np.vectorize(rrstr)
```

# 

```python
n = 16
sigma = 5
alpha = 0.05
mu0 = 20
mu1 = 24

# Критическое значение для уровня значимости
z_crit =  norm.ppf(1-alpha/2)


# Мощность теста (a)
z_power =  (mu1 - mu0) / (sigma / np.sqrt(n)) - z_crit
power = norm.cdf(z_power)

display(Math(r"\text{Мощность теста (W) при n = }"+str(n) + f': {v_rrstr(power,6)}'))
```

```python
sigma = 0.4
n=64
x_ = 3.07
alpha = 0.05
mu0 = 3



z = (x_ - mu0)/ (sigma / np.sqrt(n))

z_crit =  norm.ppf(1-alpha)

p_value = 1 - norm.cdf(z)

# Вывод результатов
display(Math(r"\text{Тестовая статистика (z): }" + str(v_rrstr(z,3)) ))
display(Math(r"\text{Критическое значение }(z_{crit}): "+ str(v_rrstr(z_crit,3))))
display(Math(r"\text{P-значение: }"+ str(v_rrstr(p_value,2))))

# Проверка гипотезы
if alpha > p_value:
    display(Math("Отвергаем нулевую гипотезу H0 в пользу альтернативной H_1.".replace(' ','~')))
else:
    display(Math("Не хватает оснований отвергнуть нулевую гипотезу H_0.".replace(' ','~')))
```

```python
sigma = 40
n = 64
x_ = 136.5
alpha = 0.01
mu0=130




z = (x_ - mu0)/ (sigma / np.sqrt(n))

z_crit_1 =  norm.ppf(1-alpha/2)
z_crit_2 =  norm.ppf(1-alpha)

p_value_1 = 2*(1 -norm.cdf(abs(z)))
p_value_2 = 1 - norm.cdf(z)

# Вывод результатов
display(Math(r"\text{Тестовая статистика (z): }" + str(v_rrstr(z,3)) ))

display(Math(r"\text{Критическое значение для двустороннего теста } (z_{crit_1}): "+ str(v_rrstr(z_crit_1,3))))
display(Math(r"\text{Критическое значение для одностороннего теста } (z_{crit_2}): "+ str(v_rrstr(z_crit_2,3))))

display(Math(r"\text{P-значение для двустороннего теста: }"+ str(v_rrstr(p_value_1,4))))
display(Math(r"\text{P-значение для одностороннего теста: }"+ str(v_rrstr(p_value_2,4))))


# Проверка гипотезы
if alpha > p_value_1:
    display(Math("a) Отвергаем нулевую гипотезу H0 в пользу альтернативной H_1.".replace(' ','~') + r'\mu \neq ' + str(mu0)))
else:
    display(Math("a) Не хватает оснований отвергнуть нулевую гипотезу H_0.".replace(' ','~')))
    
if alpha > p_value_2:
    display(Math("b) Отвергаем нулевую гипотезу H0 в пользу альтернативной H_1.".replace(' ','~') + r'\mu > ' + str(mu0)))
else:
    display(Math("b) Не хватает оснований отвергнуть нулевую гипотезу H_0.".replace(' ','~')))
```
