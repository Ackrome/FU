```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts

def rrstr(x,n=0): # округление до n знаков после запятой
    if n == 0:
        return str(x)
    fmt = '{:.'+str(n)+'f}'
    return fmt.format(x).replace('.',',')
```

```python
n = []
n.append(np.array('''13
14
15
17
16
15'''.split('\n')).astype(int))
n.append(np.array('''18
19
22
20
24
23'''.split('\n')).astype(int))

q = [i.size for i in n]
N = sum(q)

avg = np.array([i.mean() for i in n])
E_n = np.multiply(avg,q).sum()/N
var = np.array([i.var() for i in n])
Var_n = np.multiply(var,q).sum()/N
Var_n_gr =np.multiply( np.subtract(E_n,avg)**2,q).sum()/N
Var = (Var_n_gr+Var_n)
n=20
Var_x_ = Var/n * (N-n)/(N-1)

print('Всего людей = '+ str(N))
print('Групповые средние = ', np.vectorize(rrstr)(avg,3))
print('Общее среднее = ' + rrstr(E_n,3))
print('1) Внутригрупповые дисперсии = ', np.vectorize(rrstr)(var,3))
print('2) Средняя групповая дисперсия = ' + rrstr(Var_n,3))
print('3) Межгрупповая дисперсия = ' + rrstr(Var_n_gr,3))
print('4) Общая дисперсия по правилу сложения = ' + rrstr(Var_n_gr+Var_n,3))
print('Общее стандартное отклоенение по правилу сложения = ' + rrstr(Var**0.5,3))
print('Дисперсия выборки = ' + rrstr(Var_x_,3))
```

Для анализа возьмем три группы сотрудников, работающих в разных отделах компании, и их данные о заработной плате (в тысячах рублей):

$$
X_1 = \{50,55,52,48,51 \}\\
X_2 = \{60,65,62,58,64\}\\
X_3 = \{70,75,72,68,74 \}
$$
Определите:
 1) групповые и общую среднюю;
 2)  внутригрупповые и межгрупповую дисперсию,
 3)  общую дисперсию по правилу сложения. 

```python
n = []
n.append(np.array('''50,55,52,48,51'''.split(',')).astype(int))
n.append(np.array('''60,65,62,58,64'''.split(',')).astype(int))
n.append(np.array('''70,75,72,68,74'''.split(',')).astype(int))

q = [i.size for i in n]
N = sum(q)

avg = np.array([i.mean() for i in n])
E_n = np.multiply(avg,q).sum()/N
var = np.array([i.var() for i in n])
Var_n = np.multiply(var,q).sum()/N
Var_n_gr =np.multiply( np.subtract(E_n,avg)**2,q).sum()/N
Var = (Var_n_gr+Var_n)

print('Всего людей = '+ str(N))
print('Групповые средние = ', np.vectorize(rrstr)(avg,3))
print('Общее среднее = ' + rrstr(E_n,3))
print('Внутригрупповые дисперсии = ', np.vectorize(rrstr)(var,3))
print('Средняя групповая дисперсия = ' + rrstr(Var_n,3))
print('Межгрупповая дисперсия = ' + rrstr(Var_n_gr,3))
print('Общая дисперсия по правилу сложения = ' + rrstr(Var_n_gr+Var_n,3))
print('Общее стандартное отклоенение по правилу сложения = ' + rrstr(Var**0.5,3))
```
