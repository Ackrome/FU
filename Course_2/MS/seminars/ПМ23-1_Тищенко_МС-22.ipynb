```python
import numpy as np
from scipy.stats import *
from sympy import *
from IPython.display import Math
from scipy import optimize
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

def rrstr(x,n): # округление до n знаков после запятой
    fmt = '{:.'+str(n)+'f}'
    return fmt.format(x).replace('.',',')

v_rrstr = np.vectorize(rrstr)
```

```python
n = 30
mu0 = 9
mu1 = 8
sigma = 16**0.5
alpha = 0.05
sigma0 = 9**0.5
sigma1 = 4.5
chi2_obs = (n-1)*sigma**2/(sigma0**2)

p_value_1 = 2*chi2.sf(chi2_obs,df = n-1)


# Вывод результатов
display(Math(r"\text{Тестовая статистика : } \Chi^2_{obs} = " + str(v_rrstr(chi2_obs,3)) ))

display(Math(r"\text{P-значение для двустороннего теста: }"+ str(v_rrstr(p_value_1,4))))


# Проверка гипотезы
if alpha > p_value_1:
    display(Math("Отвергаем нулевую гипотезу H0 в пользу альтернативной H_1.".replace(' ','~')))
else:
    display(Math("Не хватает оснований отвергнуть нулевую гипотезу H_0.   ".replace(' ','~')))
    
    
chi2_crit_1 =  chi2.ppf(alpha/2,n-1)
chi2_crit_2 =  chi2.isf(alpha/2,n-1)

display(Math('Interval = ' + str((chi2_crit_1, chi2_crit_2))))

# Вероятность ошибки второго рода
beta = chi2.cdf(chi2_crit_2*(sigma1/(sigma0))**2,df = n-1) - chi2.cdf(chi2_crit_1*(sigma1/(sigma0))**2, df = n-1)
display(Math(f"Ошибка второго рода (β): {beta:.6f}".replace(' ','~')))
```
