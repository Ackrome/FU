```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sympy import *
from functools import reduce
from IPython.display import Math

from math import comb

def bernulli(n,i,p):
    return comb(n,i)*p**i*(1-p)**(n-i)
```

```python
display(Math('Вероятность~\mathbb{P}(\{\hat{F}(7)=\hat{F}(10)\}) = '+latex((2/3)**8)))
display(Math('Вероятность~\mathbb{P}(\{\hat{F}(8)=1/4\}) = '+latex(bernulli(8,2,2/3))))
```

```python
Omega = np.array([53, 34, 61, 58, 83, 74, 0, 87, 0, 79])
n = 5

Var = Omega.var(ddof=0)/n
mu3 = st.moment(Omega, 3)/n**2
display(Math('Дисперсия~\mathbb{Var}(\overline{X}) = ' + latex(Var)))
display(Math('Центральный~момент~\mu_3(\overline{X}) = ' + latex(mu3)))
```

```python
X = np.repeat([2,3,4,5],[7,22,49,22])
n_prep=5
display(Math('Математическое~ожидание~\mathbb{E}(\overline{X}) = ' + latex(X.mean())))
display(Math('Стандартное~отклонение~\sigma(\overline{X}) = ' + latex(X.std() * np.sqrt((X.shape[0] - X.shape[0]/n_prep)/(X.shape[0]/n_prep * (X.shape[0]-1)) ))))
```

```python
X=st.binom(10,1/2)
N = 2**10
n = 20
display(Math('Математическое~ожидание~\mathbb{E}(\overline{X}) = ' + latex(X.mean())))
display(Math('Дисперсия~\mathbb{Var}(\overline{X}) = ' + latex(X.var() / n * (N-n)/(N-1))))
```

$$
N - генеральная~совокупность\\
n - величина~выборки\\
\mathbb{E}(X_1)=\mu\\
\mathbb{Var}(X_1)=\sigma^2\\
\overline{X_n} - среднее~выборки\\
\Downarrow\\
\mathbb{E}(\overline{X_n})=\mathbb{E}(X_1)=\mu\\~\\
\mathbb{Var}(\overline{X_n})=\frac{\mathbb{Var}(X_1)}{n}=\frac{\sigma^2}{n}\\
------------------\\
n=1\\
\Downarrow\\
\mathbb{E}(\overline{X_n})=\mathbb{E}(X_1)=\mu\\~\\
\mathbb{Var}(\overline{X_n})=\mathbb{Var}(X_1)=\sigma^2\\
------------------\\
n=N\\
\Downarrow\\
\mathbb{E}(\overline{X_n})=\mathbb{E}(X_1)=\mu\\~\\
\mathbb{Var}(\overline{X_n})=\frac{\mathbb{Var}(X_1)}{N}=\frac{\sigma^2}{N}\\
------------------\\
n=\infty\\
\Downarrow\\
\mathbb{E}(\overline{X_n})=\mathbb{E}(X_1)=\mu\\~\\
\mathbb{Var}(\overline{X_n})=\frac{\mathbb{Var}(X_1)}{\infty}\approx 0
$$
