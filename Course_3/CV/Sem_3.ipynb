**Задание 1:** Учитывая следующее изображение:

pic 1.png

- Рассчитать и нарисовать гистограмму, нормализованную гистограмму (доля пикселей) изображения I.
- Вычислить и отобразить негативный образ I.
- Получить изображения и визуализировать их любым способом (например, через Matplotlib или Pillow)

```python
I = np.array([
    [3,3,2],
    [1,1,0],
    [2,2,2]
])
I
```

```python
fig,ax = plt.subplots(1,3, dpi = 100, figsize=(18, 6))
ax=ax.flatten()
ax[1].imshow(255-I, cmap='grey')
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title('Негатив')



ax[2].imshow(I, cmap='grey')
ax[2].set_xticks([])
ax[2].set_yticks([])
ax[2].set_title('Оригинал')


C = Counter(I.flatten())

ax[0].hist(C.keys(), 128, weights=np.bincount(I.flatten())/I.flatten().sum())
ax[0].set_title('Гистограмма')

plt.tight_layout()
plt.show()
```

**Задание 3:** Учитывая следующее изображение J:

pic 2.png

- Рассчитать и нарисовать гистограмму, нормализованную гистограмму изображения J.
- Преобразуйте изображение J в двоичное изображение под названием B, используя метод определения порога, где заранее определенный порог k — это пиксель, который появляется на изображении с наибольшей частотой.
- Визуализируйте полученные результаты.

```python
J = np.array([
    [0,2,1,7],
    [3,2,5,2],
    [1,1,7,6],
    [5,0,0,3]
])
```

```python
fig,ax = plt.subplots(1,3, dpi = 100, figsize=(18, 6))
ax=ax.flatten()


C = Counter(J.flatten())
ax[0].hist(C.keys(), 128, weights=np.array(list(C.values()))/sum(list(C.values())))
ax[0].set_title('Гистограмма')


J_bitnorm = (J - np.min(J))/(np.max(J) - np.min(J))
J_bitnorm[J_bitnorm<0.5] = 0
J_bitnorm[J_bitnorm>=0.5] = 1
J_bitnorm
ax[1].imshow(J_bitnorm, cmap='grey')
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title('Двоичное')



ax[2].imshow(J, cmap='grey')
ax[2].set_xticks([])
ax[2].set_yticks([])
ax[2].set_title('Оригинал')




plt.tight_layout()
plt.show()
```
