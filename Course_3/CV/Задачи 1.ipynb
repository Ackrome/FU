```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.stats import mode

img = Image.open("Noise_salt_and_pepper.png").convert('L')
img_array = np.array(img, dtype=float) / 255.0
```

# Задача 1

Напишите собственную реализацию фильтра среднего и примените его к изображению с шумом типа «соль-перец». Опишите вывод, эффективно ли фильтр среднего удаляет шум. Без OpenCV, на Python.

```python
def compute_mean(image):
    height, width = image.shape
    new_pic = np.zeros_like(image, dtype=np.float64)
    padded_image = np.pad(image, ((1, 1), (1, 1)), mode='constant', constant_values=0)

    for i in range(1, height + 1):
        for j in range(1, width + 1):
            neighborhood = padded_image[i-1:i+2, j-1:j+2]

            new_pic[i-1, j-1] = np.mean(neighborhood)
            
    return new_pic

plt.imshow(img, cmap='grey')
plt.xticks([])
plt.yticks([])
plt.show()

plt.imshow(compute_mean(img_array), cmap='grey')
plt.xticks([])
plt.yticks([])
plt.show()
```

# Задача 4

Реализуйте фильтр, который для каждого пикселя в окрестности находит наиболее часто встречающееся значение яркости (моду гистограммы окна). Проанализируйте, в каких сценариях (типы изображений, шумов) он может работать лучше или хуже медианного.


```python
def compute_mode(image):
    height, width = image.shape
    new_pic = np.zeros_like(image, dtype=np.float64)
    padded_image = np.pad(image, ((1, 1), (1, 1)), mode='constant', constant_values=0)

    for i in range(1, height + 1):
        for j in range(1, width + 1):
            neighborhood = padded_image[i-1:i+2, j-1:j+2]
            new_pic[i-1, j-1] = mode(neighborhood, axis=None, keepdims=True).mode[0]

            
    return new_pic


plt.imshow(img, cmap='grey')
plt.xticks([])
plt.yticks([])
plt.show()

plt.imshow(compute_mode(img_array), cmap='grey')
plt.xticks([])
plt.yticks([])
plt.show()
```
