````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
````

# Задание 1

````python
x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,300)

f, a = plt.subplots(2, 2,dpi=100, figsize= (10,10))

a[0, 0].scatter(range(100), x)
a[0, 1].scatter(range(300), y)

a[1, 0].hist(x)
a[1, 1].hist(y)

plt.show()
````

````python
m = np.mean(x)
v = np.var(x)

s, p = st.kstest(x, 'uniform')

print(m)
print(v)
print(p)
````

````python
m = np.mean(y)
v = np.var(y)

s, p = st.kstest(y, 'uniform')

print(m)
print(v)
print(p)
````

# Отчет

**1. Использование генератора случайных чисел (ГСЧ) в методе Монте-Карло**
Метод Монте-Карло целиком построен на многократном моделировании случайных событий для вычисления вероятностей или численного решения сложных математических задач (например, вычисления площади сложных фигур или оценки финансовых рисков). ГСЧ выступает здесь "движком": он генерирует базовые равномерно распределенные числа на интервале[0, 1), которые затем с помощью математических преобразований превращаются в любые другие распределения (нормальное, экспоненциальное и т.д.), имитируя реальные физические или бизнес-процессы. От качества равномерного ГСЧ напрямую зависит точность и достоверность всей модели Монте-Карло.

**2. Влияние количества экспериментов на качество данных**
Согласно закону больших чисел, чем больше сгенерированных значений (размер выборки), тем точнее эмпирическое распределение совпадает с теоретическим. 
*   **Малые выборки (например, 5 значений):** Данные распределены хаотично. Среднее значение и дисперсия могут сильно отклоняться от идеальных (0.5 и ~0.0833 соответственно). Гистограмма не показывает форму распределения.
*   **Большие выборки (100, 300, 500 и выше):** Столбцы гистограммы выравниваются, показывая "плоскую" форму равномерного распределения. Статистические метрики (математическое ожидание и дисперсия) стабилизируются. P-value в тесте Колмогорова-Смирнова (`kstest`) становится более надежным индикатором того, что выборка действительно принадлежит к заданному закону.

**Дополнительные пункты:**

**3. Псевдослучайность и воспроизводимость экспериментов**
Компьютерные генераторы (включая функцию `rand` в `numpy`) создают *псевдослучайные* числа. Они вычисляются по строгим математическим формулам. Если в методе Монте-Карло или при отладке кода требуется, чтобы график и значения не менялись при перезапуске программы, необходимо зафиксировать начальное состояние генератора (seed). 

```python
import numpy as np

np.random.seed(42)
x = np.random.rand(100)
```

**4. Визуальная vs Статистическая оценка качества**
Точечные диаграммы (`scatter`) полезны лишь для выявления явных паттернов или ошибок ГСЧ (например, если числа ложатся в видимые линии, что говорит о плохом генераторе). Гистограммы (`hist`) дают общее понимание формы. Однако "качество" данных для точных научных расчетов можно подтвердить только статистическими тестами (например, `scipy.stats.kstest`), которые выдают конкретное числовое значение p-value для принятия или отклонения гипотезы о равномерности.

**5. Масштабирование интервалов**
Стандартные функции генерируют значения в отрезке[0, 1). Если для моделирования требуются числа в другом диапазоне (например, [10, 70]), применяется либо масштабирование базового массива случайных величин (`x = a + (b - a) * np.random.rand(N)`), либо специальные функции библиотек (`np.random.uniform(10, 70, N)`), что делает код чище и быстрее.
