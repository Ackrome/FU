# 1.4

Сколько есть пятизначных чисел, которые одинаково читаются слева направо и справа налево (например, таких, как 67876, 17071)?

Ответ: 900.

```python
Chisla = '0123456789'

k=0
for a in Chisla:
    for b in Chisla:
        for c in Chisla:
            if a!='0':
                k+=1
                print(a+b+c+b+a)
print(f'Кол-во:{k}')
```

# 1.8

a)Число 3^5 * 5^4 имеет (5 + 1) * (4 + 1) = 30 делителей.Это происходит потому, что для каждого из 5 множителей 3^i (где 0 <= i <= 5) мы можем выбрать 0, 1, 2, 3, 4 или 5 степеней 3.Аналогично, для каждого из 4 множителей 5^j (где 0 <= j <= 4) мы можем выбрать 0, 1, 2, 3 или 4 степени 5.Перемножая эти возможности, получаем 6 * 5 = 30 делителей.

б) Следуя логике из прошлого объяснения можно утверждать, что  искомое количество делителей будет находиться по формуле

∑(αi + 1) для (i -> ∞ ) 

```python
sp.factorint(13499773954411572828000) # Ищет простые делители и их степени
```

```python
from functools import reduce
reduce((lambda a,b : (a+1)*(b+1)),list(sp.factorint(13499773954411572828000).values()))
```
