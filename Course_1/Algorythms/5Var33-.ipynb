Реализовать класс стек через использование списка:
   * Создать класс `Stack` с методами:
     - `push(item)`: добавляет элемент `item` в верхушку стека;
     - `pop()`: удаляет и возвращает элемент из верхушки стека;
     - `peek()`: возвращает элемент из верхушки стека без его удаления;
     - `is_empty()`: проверяет, пуст ли стек;
     - `size()`: возвращает текущий размер стека.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.is_empty():
            raise ValueError("Стек пуст!")
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            raise ValueError("Стек пуст!")
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

     - `pop()`: удаляет и возвращает элемент из верхушки стека;
     - `peek()`: возвращает элемент из верхушки стека без его удаления;
     - `is_empty()`: проверяет, пуст ли стек;
     - `size()`: возвращает текущий размер стека

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def push(self, item):
        new_node = Node(item)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.is_empty():
            raise ValueError("Стек пуст!")
        temp = self.top
        self.top = self.top.next
        return temp.data

    def peek(self):
        if self.is_empty():
            raise ValueError("Стек пуст!")
        return self.top.data

    def is_empty(self):
        return self.top is None

    def size(self):
        current = self.top
        count = 0
        while current is not None:
            count += 1
            current = current.next
        return count
```

   * Создать класс `Queue` с методами:
     - `enqueue(item)`: добавляет элемент `item` в конец очереди;
     - `dequeue()`: удаляет и возвращает элемент из начала очереди;
     - `peek()`: возвращает элемент из начала очереди без его удаления;
     - `is_empty()`: проверяет, пуста ли очередь;
     - `size()`: возвращает текущий размер очереди.

```python
class Queue:
    def __init__(self):
        self._items = []

    def enqueue(self, item):
        self._items.append(item)

    def dequeue(self):
        if self.is_empty():
            return None
        return self._items.pop(0)

    def peek(self):
        if self.is_empty():
            return None
        return self._items[0]

    def is_empty(self):
        return len(self._items) == 0

    def size(self):
        return len(self._items)


# Пример использования
queue = Queue()

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print(queue.size())  # 3

print(queue.dequeue())  # 1

print(queue.peek())  # 2

print(queue.is_empty())  # False

queue.dequeue()
queue.dequeue()

print(queue.is_empty())  # True
```

```python
class Queue:
    def __init__(self):
        self.head = None
        self.tail = None

    def enqueue(self, item):
        new_node = Node(item)
        if self.tail is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        return item

    def dequeue(self):
        if self.is_empty():
            return None
        node = self.head
        self.head = self.head.next
        if self.head is None:
            self.tail = None
        return node.data

    def peek(self):
        if self.is_empty():
            return None
        return self.head.data

    def is_empty(self):
        return self.head is None

    def size(self):
        node = self.head
        count = 0
        while node is not None:
            count += 1
            node = node.next
        return count
    
    def sort(self):
        '''Сортировка пузырьком'''
        
        if self.head is None:
            return
        current = self.head
        while current.next is not None:
            next_node = current.next
            if current.data >= next_node.data:
                current.data, next_node.data = next_node.data, current.data
            current = current.next
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_at_head(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            self.tail.next = new_node

    def insert_at_tail(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            new_node.next = self.head
            self.head.prev = new_node
            self.tail = new_node

    def get_length(self):
        if self.head is None:
            return 0

        length = 1
        current = self.head.next
        while current != self.head:
            length += 1
            current = current.next

        return length

    def delete_at_head(self):
        if self.head is None:
            return

        if self.head == self.tail:
            self.head = None
            self.tail = None
            return

        self.head = self.head.next
        self.head.prev = self.tail
        self.tail.next = self.head

    def delete_at_tail(self):
        if self.head is None:
            return

        if self.head == self.tail:
            self.head = None
            self.tail = None
            return

        prev = self.tail.prev
        prev.next = self.head
        self.head.prev = prev
        self.tail = prev

    def print_list(self):
        if self.head is None:
            print("Список пуст")
            return

        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next

            if current == self.head:
                break
    def sort(self):
        if self.head is None:
            print("Список пуст")
            return

        current = self.head
        elements=[]
        while current:
            elements.append(current.data)
            current = current.next
            
            if current == self.head:
                break
        length = self.get_length()
        sorted_elems = sorted(elements)
        
        for i in range(length):
            self.delete_at_head()
        

        for i in range(length):
            self.insert_at_tail(sorted_elems[i])
        self.print_list()
        

# Пример использования
dll = CircularDoublyLinkedList()

dll.insert_at_head(10)
dll.insert_at_head(20)
dll.insert_at_head(30)
dll.insert_at_tail(60)
dll.insert_at_tail(40)
dll.insert_at_tail(50)
dll.insert_at_tail(50)

print("Список:")
dll.print_list()

print("\nДлина списка:", dll.get_length())

dll.delete_at_head()
print("\nСписок после удаления элемента из головы:")
dll.print_list()

dll.delete_at_tail()
print("\nСписок после удаления элемента из хвоста:")
dll.print_list()


print("\nСписок после сортировки:")
dll.sort()
```

Необходимо отсортировать массив строк по длине и вывести результат на экран. В зависимости от переданного параметра отсортировать массив строк по возрастанию длины или по убыванию длины, используя алгоритмы сортировки: сортировку выбором, сортировку пузырьком и быструю сортировку. Сравнить время выполнения алгоритмов сортировки с помощью декоратора. Строки хранятся в файле.

```python
import random
import string
from tqdm.notebook import tqdm

def selection_sort(array):
    """
    Сортирует массив целых чисел с помощью сортировки выбором.

    Args:
        array (list): Исходный массив.

    Returns:
        list: Отсортированный массив.
    """
    n = len(array)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if array[j][0] < array[min_idx][0]:
                min_idx = j
        array[i], array[min_idx] = array[min_idx], array[i]
    return array

def bubble_sort(array):
    """
    Сортирует массив целых чисел с помощью сортировки пузырьком.

    Args:
        array (list): Исходный массив.

    Returns:
        list: Отсортированный массив.
    """
    n = len(array)
    for i in range(n):
        for j in range(n - i - 1):
            if array[j][0] > array[j + 1][0]:
                array[j], array[j + 1] = array[j+1], array[j]
    return array


def quick_sort(array):
    """
    Сортирует массив целых чисел с помощью быстрой сортировки.

    Args:
        array (list): Исходный массив.
        start_index (int): Индекс начала подмассива (по умолчанию 0).
        end_index (int): Индекс конца подмассива (по умолчанию длина массива - 1).

    Returns:
        list: Отсортированный массив.
    """
    if len(array) <= 1:
        return array
    else:
        pivot = array[0]
        left = []
        right = []
        for i in range(1, len(array)):
            if array[i][0] < pivot[0]:
                left.append(array[i])
            else:
                right.append(array[i])
        return quick_sort(left) + [pivot] + quick_sort(right)

def sort(array,algorythm='quick sort', reverse=False):
    array =  [(len(x),x) for x in array]
    new_array = list(map(lambda x: x[1],({'quick sort': quick_sort,'bubble sort':bubble_sort,'selection sort':selection_sort}[algorythm])(array)))
    return new_array[::-1] if reverse else new_array 

def generate_random_string(file_name:str,min_length:int, max_length:int,n_strings:int):
    """
    Генерирует случайную строку заданной длины.

    Args:
        min_length (int): Минимальная длина строки.
        max_length (int): Максимальная длина строки.
        n_strings (int): Количество генерируемых строк в массив

    Returns:
        nd.array: массив строк
    """
    if min_length < 1 or max_length < min_length:
        raise ValueError("min_length must be less than or equal to max_length")
    
    sis = []
    for i in range(n_strings):
        sis.append(''.join(random.choice(string.ascii_letters + string.digits + string.punctuation) for _ in range(random.randint(min_length, max_length))))

    with open(file_name, 'w') as f:
        for i in range(n_strings):
            f.write(sis[i]+'\n')
    return


generate_random_string('test.txt', 1, 300,10000)
```

Реализовать класс бинарного дерева. Написать функцию для нахождения наибольшей глубины листьев в бинарном дереве.

```python
class Node:
  """
  Узел бинарного дерева.
  """

  def __init__(self, value):
    self.value = value
    self.left = None
    self.right = None


class BinaryTree:
  """
  Класс бинарного дерева.
  """

  def __init__(self):
    self.root = None

  def insert(self, value):
    """
    Вставляет значение в бинарное дерево.
    """
    new_node = Node(value)
    if self.root is None:
      self.root = new_node
      return

    parent = None
    current = self.root
    while current is not None:
      parent = current
      if value < current.value:
        current = current.left
      else:
        current = current.right

    if value < parent.value:
      parent.left = new_node
    else:
      parent.right = new_node

  def find_max_depth(self, node=None):
    """
    Находит максимальную глубину листьев в бинарном дереве.
    """
    if node is None:
      node = self.root

    if node.left is None and node.right is None:
      return 1

    left_depth = self.find_max_depth(node.left)
    right_depth = self.find_max_depth(node.right)

    return max(left_depth, right_depth) + 1

  def print_tree(self, node=None):
    """
    Печатает бинарное дерево с ветвями.
    """
    if node is None:
      node = self.root

    if node is None:
      print("Дерево пустое")
      return

    print(node.value)
    self.print_tree_helper(node.left, "  |")
    self.print_tree_helper(node.right, "  |")

  def print_tree_helper(self, node, indent):
    """
    Вспомогательная функция для печати бинарного дерева с ветвями.
    """
    if node is None:
      return

    print(indent + str(node.value))
    self.print_tree_helper(node.left, indent + " |")
    self.print_tree_helper(node.right, indent + " |")
    
tree = BinaryTree()

tree.insert(5)
tree.insert(3)
tree.insert(7)
tree.insert(2)
tree.insert(4)
tree.insert(6)
tree.insert(8)

print(f"Максимальная глубина листьев: {tree.find_max_depth()}")
tree.print_tree()
```

Создайте структуру данных «Top N», которая хранит $N$ наибольших элементов из потока входных данных. Структура должна поддерживать операцию добавления нового элемента в поток и операцию извлечения одного из «Top N» элементов. Используйте для этого двоичную кучу.

```python
class TopN:
  """
  Структура данных "Top N", основанная на двоичной куче.
  """

  def __init__(self, n):
    """
    Инициализирует структуру данных "Top N".

    Args:
      n: Максимальное количество элементов.
    """
    self.n = n
    self.heap = []  # Двоичная куча для хранения элементов

  def insert(self, item):
    """
    Добавляет новый элемент в поток входных данных.

    Args:
      item: Элемент, который нужно добавить.
    """
    if len(self.heap) < self.n:
      # Если куча не заполнена, добавляем элемент напрямую
      self.heap.append(item)
      self.heapify_up(len(self.heap) - 1)
    else:
      # Если куча заполнена, сравниваем новый элемент с минимальным
      # элементом в куче. Если новый элемент больше, заменяем
      # минимальный элемент новым.
      if item > self.heap[0]:
        self.heap[0] = item
        self.heapify_down(0)

  def get_top_n(self):
    """
    Возвращает список из $N$ наибольших элементов.

    Returns:
      Список $N$ наибольших элементов.
    """
    top_n = []
    for i in range(self.n):
      top_n.append(self.heap[i])
    return top_n

  def heapify_up(self, index):
    """
    Поднимает элемент в куче до его правильного места.

    Args:
      index: Индекс элемента в куче.
    """
    parent = (index - 1) // 2
    while index > 0 and self.heap[index] > self.heap[parent]:
      self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
      index = parent
      parent = (parent - 1) // 2

  def heapify_down(self, index):
    """
    Опускает элемент в куче до его правильного места.

    Args:
      index: Индекс элемента в куче.
    """
    left_child = 2 * index + 1
    right_child = 2 * index + 2
    largest = index
    if left_child < len(self.heap) and self.heap[left_child] > self.heap[largest]:
      largest = left_child
    if right_child < len(self.heap) and self.heap[right_child] > self.heap[largest]:
      largest = right_child
    if largest != index:
      self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index]
      self.heapify_down(largest)

# Пример использования
top_n = TopN(3)

top_n.insert(10)
top_n.insert(5)
top_n.insert(15)
top_n.insert(2)
top_n.insert(1)
print(top_n.get_top_n())
```
